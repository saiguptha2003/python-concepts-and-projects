```python
sum=0
count=0
for i in dataframe.columns:
    if("Revenue" in i and "Budget" in i):
        count=count+1
        total=dataframe.loc[15][i]
        if(type(total)==str):
            total=float(total)
        print(i,total,sep="=")
        sum=total+sum
print("Average budget of anual 2021-2022 and 2022-2023 is ",sum//count)
    
```
